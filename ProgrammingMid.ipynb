```python
#Q2
import re
df3 = pd.read_csv("/content/Motor_Vehicle_Collisions_-_Crashes.csv")
df3
```

```python

df3['CRASH DATE'].apply(lambda x: re.search('^\d{2}', x).group(0))
```

```python
def Nan_dealer(x):
  if x == None:
    return ".0"
  return x.group(0)
df3['LATITUDE'].apply(lambda x: Nan_dealer(re.search('\.\d*$', str(x))))
```

```python
df3['LONGITUDE'].apply(lambda x: Nan_dealer(re.search('\.\d*$', str(x))))
```

```python
df3['Regex'] = df3['CRASH DATE'].apply(lambda x: re.search('^\d{2}', x).group(0)) + df3['LATITUDE'].apply(lambda x: Nan_dealer(re.search('\.\d*$', str(x)))) + df3['LONGITUDE'].apply(lambda x: Nan_dealer(re.search('\.\d*$', str(x))))
df3.head(500).to_csv("RegexCarCollision.csv")
```
